```python
#Job Deatails(Comparing total job openings with Total Applicants in each company)
Job_Details = df.groupby("Company Name").agg(Total_Job_Openings=("Company Name", "count"), Total_Applicants=("Total_applicants", "sum")).reset_index()
Job_Details
```

```python
companies = Job_Details["Company Name"]
job_openings = Job_Details["Total_Job_Openings"]
total_applicants = Job_Details["Total_Applicants"]

# Define the x-axis positions
x = np.arange(len(companies))

# Bar width
bar_width = 0.4

# Create a grouped bar chart
plt.figure(figsize=(16, 8), dpi=150)
barL1=plt.bar(x - bar_width/2, job_openings, width=bar_width, label="job_openings", color="skyblue")
barL2=plt.bar(x + bar_width/2, total_applicants, width=bar_width, label="total_applicants", color="orange")

plt.bar_label(barL1, label_type="edge", fontsize=10)
plt.bar_label(barL2, label_type="edge", fontsize=10)

# Add labels and title
plt.xlabel("Company Name", fontsize=14)
plt.ylabel("Counts", fontsize=14)
plt.title("Job Openings and Total Applicants by Company", fontsize=16)
plt.xticks(x, companies, rotation=80, fontsize=12)
plt.legend(fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Creating a DataFrame for skills count
skills = df.drop(["Total Job Openings","Company Name", "Designation", "Location", "Total_applicants", "LinkedIn_Followers", "Involvement", "Level", "Employee_count","Industry"], axis=1)
# Initializing an empty dictionary 
dic_skills = {}

# Counting the number of jobs requiring each skill
for i in skills.columns:
    dic_skills[i] = df[i].sum()

# Converting the dictionary to a list of tuples
dt = list(dic_skills.items())

# Creating a DataFrame from the skill count data
df_skill = pd.DataFrame(dt, columns=['Skill', 'Number of jobs requiring the skill'])
df_skill
```

```python
plt.figure(figsize=(20,10),dpi=100)
chart=sns.barplot(x=df_skill["Skill"],y=df_skill["Number of jobs requiring the skill"],palette="inferno")
plt.title("Number of jobs that require a particular skill",size=25)
plt.ylabel("Number of jobs",size=22)
plt.xlabel("Skill",size=22)
plt.xticks(fontsize=18,rotation=90)
plt.yticks(fontsize=18)
chart.bar_label(chart.containers[0],size=18)
```
